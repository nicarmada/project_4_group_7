```python
import pandas as pd

# Load the main dataset
data_path = 'data/Multiple_Cause_of_Death,_1999-2014_v1.1.csv'
df = pd.read_csv(data_path)

# Inspect dataset structure
print(df.info())
print(df.head())

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)

# Extract list of states before cleaning
original_states = set(df["State"].unique())  # Capture all unique state names

# Save cleaned dataset for later steps
cleaned_data_path = 'data/cleaned_multiple_cause_of_death.csv'
df.to_csv(cleaned_data_path, index=False)

# Load cleaned dataset to compare states
df_cleaned = pd.read_csv(cleaned_data_path)
cleaned_states = set(df_cleaned["State"].unique())  # Extract states after cleaning

# Identify missing states
missing_states = original_states - cleaned_states

# Print missing states if any
print("\nChecking for missing states after data cleaning...")
if missing_states:
    print("States missing after cleaning:")
    print(", ".join(sorted(missing_states)))  # Sorted for readability
else:
    print("No states were lost during cleaning.")

print("\nData cleaning completed. Cleaned dataset saved successfully.")
```
